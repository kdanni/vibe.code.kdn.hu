# Run prompt helper from a notebook
This notebook demonstrates how to reuse the script tooling with helpful defaults.
Set a few environment variables (or override via code), ping the service, and preview a prompt file.

```python
import os
import sys
from pathlib import Path

# Make sure the repository root is importable for the scripts package
repo_root = Path().resolve()
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

from scripts.bootstrap import add_repo_root_to_path
from scripts.run_prompt import run_prompt, usage_hint

# Ensure imports work even if the package is not installed globally
add_repo_root_to_path()

# Provide friendly default values for local testing
os.environ.setdefault('LLM_API_BASE_URL', 'http://localhost:8000')
os.environ.setdefault('LLM_API_KEY', 'dev-key')

print(usage_hint())

# Preview the prompt and (optionally) ping the service
run_prompt('prompts/prompt_template.md')
```
